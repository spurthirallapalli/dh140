## **Step 2: Shakespeare word frequency**

For this assigment, I will analyze the word frequency in *Romeo and Juliet*. I found the plain text UTF-8 file on the Projet Gutenberg website at this link: https://www.gutenberg.org/cache/epub/1513/pg1513.txt. I will be using requests and Beautiful Soup to get the text.

```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.gutenberg.org/cache/epub/1513/pg1513.txt')
response
```

```python
# Take a quick look at the text to make sure it's correct

response.text
```

```python
# Reformat using Beautiful Soup
html_string = response.text
document = BeautifulSoup(html_string, "html.parser")
document
```

Now, tokenize the words and remove stopwords.

```python
# Imports required for tokenization and stopword removal
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
```

```python
# There are a lot of lines that are extra here, but I don't predict that they will be too important to the outcome
# But, I will add Gutenberg to the list of stop words since it is mentioned quite a few times
# Adding a word to a list using append method
# Originally tried to add list("Gutenberg") but that split up the word in the list to individual letters

myStopWords = list(punctuation) + stopwords.words('english')
myStopWords.append("Gutenberg")
```

```python
print(myStopWords)
```

```python
# We need to tokenize the script for Romeo and Juliet now

# First sentence tokenization
sent = sent_tokenize(document)

# Then word tokenization
words = []
for s in sent:
    for w in word_tokenize(s):
        words.append(w)
        
print(words)
```

```python
# Let's try this again. I think I put in the BeautifulSoup version in the sent_tokenization module, which is wrong
# So I tried the html_string (ie, response.txt) to see if this works

# First sentence tokenization
sent = sent_tokenize(html_string)

# Then word tokenization
words = []
for s in sent:
    for w in word_tokenize(s):
        words.append(w)
        
print(words)
```

```python
# This version seems to work!
# Now, I will use list comprehension like described in class to remove the stop words

wordsNoStop = [w for w in words if w not in myStopWords]
```

```python
# Now we need to find the top 20 most frequent words

from nltk.probability import FreqDist
```

```python
freq = FreqDist(wordsNoStop)
```

```python
for i in sorted(freq, key=freq.get, reverse=True)[:20]:
    print(i,freq[i])
```

A first analysis of this shows that I didn't quite do the amount of cleaning of this text that might be required to give a more accurate result, so I'm going to attempt some parts of these again/ 

Things I'm looking to change: 
- lowercase all words so that ROMEO and Romeo don't get counted as separate words
- add stopwords: ’

```python
myStopWordsNew = list(punctuation) + stopwords.words('english')
myStopWordsNew.append("Gutenberg")
myStopWordsNew.append("’")
```

```python
# We already have all the necessary words in the list "words", we just need to lowercase them all
# Not sure if we covered this in class, but here's a link I found: 
# https://www.kite.com/python/answers/how-to-make-the-elements-in-a-list-of-strings-lowercase-in-python

wordsLower = [each_string.lower() for each_string in words]
```

```python
wordsNoStopNew = [w for w in wordsLower if w not in myStopWordsNew]
```

```python
# Now let's try this again

freqNew = FreqDist(wordsNoStopNew)
for i in sorted(freqNew, key=freqNew.get, reverse=True)[:20]:
    print(i,freqNew[i])
```

This seems like a more accurate and better list! Right off the bat, we can see that the words Romeo and Juliet are in the top 3 most frequently seen words. This makes sense, as they are the two main characters of the story. The top 20 words in general have a lot of names of people who are essential to the plot of the play (capulet, friar, mercutio, lawrence, tybalt, benviolio, etc). There were also some words that I noticed when I was reading the play that I did see here as well, such as "thou", "thy", and "love", so this analysis seems pretty consistent with what I observed!

```python
from nltk.stem.porter import PorterStemmer
p_stemmer = PorterStemmer()
```

```python
listOfStemmedWords = []
for i in listOfMoviesWords:
    listOfStemmedWords.append([p_stemmer.stem(w) for w in i])
```

```python
listOfStemmedWords[0]
```

```python
!pip install gensim
```

```python
from gensim import corpora, models
import gensim
```

```python
dictionary = corpora.Dictionary(listOfStemmedWords)
```

```python
corpus = [dictionary.doc2bow(text) for text in listOfStemmedWords]
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, 
                                           num_topics=2, 
                                           id2word = dictionary, 
                                           passes=20)
```

```python
for i in ldamodel.print_topics(num_topics=2, num_words=20):
    print(i)
```
