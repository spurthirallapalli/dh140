Now, I want to try implementing a **mathematical operation**. For this very first data value (shown above), I will multiply the depth value by 1000 to change the units from km to m.

Note: According to [this webpage](https://www.usgs.gov/programs/earthquake-hazards/determining-depth-earthquake), the depth of an earthquake is measured in kilometers. 

```python
# Using mathematical operations

#modified: earthquake_data[0]['location']['depth'] -> earthquakes.loc[0, 'location.depth']
depth_meters = earthquakes.loc[0, 'location.depth'] * 1000
print("Depth in meters is: " + str(depth_meters) + " meters.")
```

This matches last weeks assignment, which produced the same result.

Now, I'm going to experiment with **conditional expressions**. According to [this website](https://www.usgs.gov/programs/earthquake-hazards/determining-depth-earthquake), also listed in the mathematical operations section, mentions that "Shallow earthquakes are between 0 and 70 km deep; intermediate earthquakes, 70 - 300 km deep; and deep earthquakes, 300 - 700 km deep", which I am going to apply to the first data value.

```python
#use if statement and conditionals, using the following information
# 0->70 km = Shallow
# 70->300 km = Intermediate
# 300->700 k = Deep

#modified: earthquake_data[0]['location']['depth'] -> earthquakes.loc[0, 'location.depth']

if (earthquakes.loc[0, 'location.depth'] < 70):
    print("This is a shallow earthquake.")
elif (earthquakes.loc[0, 'location.depth'] < 300):
    print("This is an intermediate earthquake")
else:
    print("This is an deep earthquake")
```

This matches last weeks assignment, which produced the same result.

In the following section, I'm going to use loc/iloc to find all the deep earthquakes and plot their values. I don't think for loops are needed here because loc/iloc takes care of that.

```python
earthquakes.loc[earthquakes['location.depth'] > 300]
```

From a quick glance above, this seemed to work!

Now, I'm going to plot these values, first with a scatter plot.

```python
# First, I'll save these values in a different variable
deepquakes = earthquakes.loc[earthquakes['location.depth'] > 300]
deepquakes
```

```python
# I want to see the correlation between magnitude and depth, so I'll plot x as magnitude and y as depth

deepquakes.plot(kind='scatter', x='impact.magnitude', y='location.depth')
```

From looking above, it doesnt seem like there's a huge correlation between magnitude and depth.
I'll try adding some more visual features to make this clear.

```python
# I also want to change the color of the bars to reflect different ranges of depths

dotcolors = []
for i in deepquakes.index:
    if deepquakes.loc[i,'location.depth'] < 300:
        dotcolors.append('grey')
    elif deepquakes.loc[i,'location.depth'] >= 300 and deepquakes.loc[i,'location.depth'] < 400:
        dotcolors.append('blue')
    elif deepquakes.loc[i,'location.depth'] >= 400 and deepquakes.loc[i,'location.depth'] < 500:
        dotcolors.append('purple') 
    elif deepquakes.loc[i,'location.depth'] >= 500 and deepquakes.loc[i,'location.depth'] < 600:
        dotcolors.append('orange')
    else:
        dotcolors.append('red')

deepquakes.plot(kind='scatter', x='impact.magnitude', y='location.depth', color=dotcolors)
```

As we can see above, there does not seem to be any linear correlation between depth and magnitude. Let's also try this with a bar graph.

```python
# I also want to change the color of the bars to reflect different ranges of depths. 
# We also want to sort the magnitudes (sort values) and make the chart bigger (figsize) for better visibility 

barcolors = []
for i in deepquakes.index:
    if deepquakes.loc[i,'location.depth'] < 300:
        barcolors.append('grey')
    elif deepquakes.loc[i,'location.depth'] >= 300 and deepquakes.loc[i,'location.depth'] < 400:
        barcolors.append('blue')
    elif deepquakes.loc[i,'location.depth'] >= 400 and deepquakes.loc[i,'location.depth'] < 500:
        barcolors.append('purple') 
    elif deepquakes.loc[i,'location.depth'] >= 500 and deepquakes.loc[i,'location.depth'] < 600:
        barcolors.append('orange')
    else:
        barcolors.append('red')

deepquakes.sort_values(by='impact.magnitude').plot(kind='bar', x='impact.magnitude', y='location.depth', color=barcolors, figsize=(12,5))
```
