### **Step 3: Write a function that takes an element from the list of CORGIS datasets, searches the respective CORGIS page for the CSV download link, and returns a Pandas dataframe.**

Like what the assignment suggested, I'm going to define a function called pdcorgis(), and then I will do the following steps.

1) Convert the dataset name to lowercase
2) Use requests with "https://corgis-edu.github.io/corgis/csv/" plus the dataset name plus "/" to get the website's HTML
3) Use BeautifulSoup to search for all the HTML link elements on that page (links have the tag "a")
4) For any link element "i", check if i.has_attr('download') is true, and then get part of the download link for the csv file via i['href']
5) The complete download link for the file will be "https://corgis-edu.github.io/corgis/csv/" plus "classics" plus "/" plus the result you found for i['href']
6) Use return corgisdf at the end of your function, where corgisdf is assigned to be the value returned by pd.read_csv() with the download link you have found

```python
import pandas as pd

def pdcorgis(name):
    # 1) Convert the dataset name to lowercase
    name = name.lower()
    # 2) Use requests with "https://corgis-edu.github.io/corgis/csv/" plus the dataset name plus "/" to get the website's HTML
    website_url = "https://corgis-edu.github.io/corgis/csv/" + name + "/"
    response = requests.get(website_url)
    html_string = response.text
    # 3) Use BeautifulSoup to search for all the HTML link elements on that page (links have the tag "a")
    document = BeautifulSoup(html_string, "html.parser")
    for i in document.find_all("a"):
        # 4) For any link element "i", check if i.has_attr('download') is true
        if i.has_attr("download") == True:
            # then get part of the download link for the csv file via i['href']
            download_link = i['href']
            #5) Create the complete link
            full_link = "https://corgis-edu.github.io/corgis/csv/" + name + "/" + download_link
    # 6) return corgisdf, assigned to be the value returned by pd.read_csv() with the download link
    corgisdf = pd.read_csv(full_link)
    return corgisdf
```
