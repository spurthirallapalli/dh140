## Mathematical Operation

Now, I want to try implementing a **mathematical operation**. For this very first data value (shown above), I will multiply the depth value by 1000 to change the units from km to m.

Note: According to [this webpage](https://www.usgs.gov/programs/earthquake-hazards/determining-depth-earthquake), the depth of an earthquake is measured in kilometers. 

```python
depth_meters = earthquake_data[0]['location']['depth'] * 1000
print("Depth in meters is: " + str(depth_meters) + " meters.")
```

*So we see using a mathematic operation that the depth of this specific earthquake is 15120 meters, or 15.12 km.*

Now, I'm going to experiment with **conditional expressions**. According to [this website](https://www.usgs.gov/programs/earthquake-hazards/determining-depth-earthquake), also listed in the mathematical operations section, mentions that "Shallow earthquakes are between 0 and 70 km deep; intermediate earthquakes, 70 - 300 km deep; and deep earthquakes, 300 - 700 km deep", which I am going to apply to the first data value.

```python
#use if statement and conditionals, using the following information
# 0->70 km = Shallow
# 70->300 km = Intermediate
# 300->700 k = Deep

if (earthquake_data[0]['location']['depth'] < 70):
    print("This is a shallow earthquake.")
elif (earthquake_data[0]['location']['depth'] < 300):
    print("This is an intermediate earthquake")
else:
    print("This is an deep earthquake")
```

*As predicted, the above printed that the earthquake we were looking at was a shallow earthquake, because it is only 15.12 km deep.*

## For Loop

In the following section, I'm going to use a **for loop** to state the type of every earthquake in the dataset.

```python
#look through every piece of earthquake data... should produce 8394 lines of classifications.

for i in earthquake_data:
    if (i['location']['depth'] < 70):
        print("This is a shallow earthquake.")
    elif (i['location']['depth'] < 300):
        print("This is an intermediate earthquake")
    else:
        print("This is an deep earthquake")
```

*As we can see printed above, there are a lot of shallow and some intermediate earthquakes in this dataset*

## Functions

I'm going to implement something similar to the for loop functionality above, but using a **function**. In this function, I will only print if an earthquake is an intermediate earthquake, along with it's location (full), and its depth. 

```python
def print_intermediate():
    for i in earthquake_data:
        if (i['location']['depth'] < 300 and i['location']['depth'] >= 70):
            print("This is an intermediate earthquake with depth " + str(i['location']['depth']) + " located at " + i['location']['full'])
```

```python
print_intermediate()
```
